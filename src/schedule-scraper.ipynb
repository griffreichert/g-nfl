```python
def generate_timestamp(row):
    timestamp = pd.to_timedelta(row['time'] + ':00')    # append :00 for seconds
    hour = int(row['time'].split(':')[0])
    if hour <= 8:        # handle international games with a rule that if the hour is greater than 8 it is morning
        timestamp += pd.Timedelta(hours=12)
    return row['date'] + timestamp

schedule['gametime'] = schedule.apply(generate_timestamp, axis=1)
```
